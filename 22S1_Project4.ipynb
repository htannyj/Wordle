```python
import tkinter as tk # code for drawing form in tkinter
from tkinter import messagebox # using it to display messagess
import random # choosing a random puzzle to guess
import json # convert python objects (dictionary) to json objects

WORD_LENGTH = 5 # a constan known as WORD_LENGTH

wordle={ # the main data structure that store game information
    'num_guess' : 0,    # number of current guesses
    'puzzle' : '',      # the puzzle
    'valid_guess' : []  # list of valid guesses 
} # wordle as a dictionary

labels = [] # The grid consists of 6 x 5 label widgets as labels list


def init_wordle(f1 = 'wordle_words.txt', f2 = 'wordle_valid_guess.txt'):
    #'''Initilaise the game. Pick the puzzle and initialise a list of valid guesses'''
    # initialise the game
    global wordle_words, wordle_valid_guess
    #wordle_words = [] # setup an empty wordle_words list
    #wordle_valid_guess = [] # setup an empty wordle_valid_guess list
    
    # open 'wordle_words.txt' file that contains wordle words list
    with open(f1, 'r') as target_words:
        # ignore the header row
        wordle_words = target_words.read().splitlines() [1:]
        #print(wordle_words[1])
        
     # open 'wordle_valid_guess.txt' file that contains wordle valid guess list
    with open(f2, 'r') as valid_guess_words:
        # ignore the header row
        wordle_valid_guess = valid_guess_words.read().splitlines() [1:]
        #print(wordle_valid_guess[1])
        
     # choose a random word as puzzle
    wordle['puzzle'] = random.choice(wordle_words)
    print("The puzzel is ", wordle['puzzle'])
    
    return wordle_words, wordle_valid_guess
    

def on_enter(event):
    #'''main logic of the game, compare the player's guess against the puzzle'''
    global wordle, labels, guess
    
    # INSERT YOUR CODE BELOW
    # Get the guess word from the one you type in the entry and save it into the guess variable
    guess = entry.get()
    print("The guess word is ", guess)
    
           
    # Check if the guess word in the wordle_valid_guess lists
    for i in range(len(wordle_valid_guess)-1):
        selected = wordle_valid_guess[i]
        if guess == selected:
            wordle['valid_guess'] = selected
            print(wordle['valid_guess'])
            messagebox.showinfo("Valid Guess.........", "Your guess word is inside Valid Guess Lists")
            break
        
        else:
            if len(guess) != WORD_LENGTH:
                break
                
    else: # after checking the guess word in the valid guess lists and cannot find it!                             
        messagebox.showerror("Not in Valid Guess", "Your guess word is not one of valid guess lists! Please type another guess.")
        wordle['valid_guess'] = []
        return
    
    #print('Continue to compare the puzzle against the guess')
    #print(wordle['num_guess'])
        
    if wordle['num_guess'] <= 5:
        
        if len(guess) == WORD_LENGTH:
            
            if wordle['puzzle'] == guess: # CORRECT
                
                puzzle = wordle['puzzle']
                
                for i, letter in enumerate(guess): # enumerate() adds a counter to an iterable and returns it in a form of enumerating object
                    
                    letter = puzzle[i]
                    label = tk.Label(win, text=letter.upper())
                    label.grid(row=wordle['num_guess'], column=i, sticky='nesw')
                    
                    label.config(font=("Arial Bold", 20), fg='WHITE', bg='GREEN', borderwidth=1, relief="solid") 
                                        
                    
                messagebox.showinfo("correct!", f"correct! the word was {wordle['puzzle']}")
                
                wordle['num_guess'] += 1
                
                answer = messagebox.askyesno(
                      title='Another game',
                      message='Play again?'
                      )

                if answer == True:
                    #print("Start a new game")
                    messagebox.showinfo("Start a new game", "You have to guess a new puzzle again")
                    init_wordle() # choose a new random puzzle
                    wordle['num_guess'] = 0
                    #labels = []
                    #win.start()
                    #win = tk.Tk()
                    #pass
                    # The grid consists of 6 x 5 label widgets
                    for r in range(6):
                        tk.Grid.rowconfigure(win, r, weight=1)
                        l = []
                        for c in range(5):
                            tk.Grid.columnconfigure(win, c, weight=1)
                            lab = tk.Label(text='', font=("Arial Bold", 20), fg='white', bg='white',  borderwidth=1, relief="solid")
                            lab.grid(row=r, column=c, sticky='nesw')
                            l.append(lab)
                        labels.append(l)
            
                elif answer == False:
                    #print("End the game")
                    messagebox.showinfo("End the game", "The grid window and game will be closed! Thank you for playing...")
                    win.destroy()           
                
                
            
            else: # INCORRECT
                puzzle = wordle['puzzle']
                for i, letter in enumerate(guess): # enumerate() adds a counter to an iterable and returns it in a form of enumerating object
                    
                    label = tk.Label(win, text=letter.upper())
                    label.grid(row=wordle['num_guess'], column=i, sticky='nesw')                     
                    
                    if letter == puzzle[i]: #if they get the letter right
                        label.config(font=("Arial Bold", 20), fg='WHITE', bg='GREEN', borderwidth=1, relief="solid") 

                    if letter in puzzle and not letter == puzzle[i]: #if the letter is in the word, but not in the right spot
                        label.config(font=("Arial Bold", 20), fg='WHITE', bg='YELLOW', borderwidth=1, relief="solid") 
                    
                    if letter not in puzzle:
                        label.config(font=("Arial Bold", 20), fg='WHITE', bg='GREY', borderwidth=1, relief="solid") 
                        
                wordle['num_guess'] += 1
          
        else:
            messagebox.showerror("please use 5 characters", "please use 5 characters in your guess")
                                     
    else:
        messagebox.showerror("you lose!", f"You Lose! The word was {wordle['puzzle']}")
        
        answer = messagebox.askyesno(
                 title='Another game',
                 message='Play again?'
                 )

        if answer == True:
                    #print("Start a new game")
                    messagebox.showinfo("Start a new game", "You have to guess a new puzzle again")
                    init_wordle() # choose a new random puzzle
                    wordle['num_guess'] = 0
                    #labels = []
                    #win.start()
                    #win = tk.Tk()
                    #pass
                    # The grid consists of 6 x 5 label widgets
                    for r in range(6):
                        tk.Grid.rowconfigure(win, r, weight=1)
                        l = []
                        for c in range(5):
                            tk.Grid.columnconfigure(win, c, weight=1)
                            lab = tk.Label(text='', font=("Arial Bold", 20), fg='white', bg='white',  borderwidth=1, relief="solid")
                            lab.grid(row=r, column=c, sticky='nesw')
                            l.append(lab)
                        labels.append(l)
            
        elif answer == False:
            #print("End the game")
            messagebox.showinfo("End the game", "The grid window and game will be closed! Thank you for playing...")
            win.destroy()
        
        
    
    return wordle, labels, guess   
                 

# === MAIN PROGRAM ===
# Initiliase the game and UI widgets, you do not need to modify the code below

init_wordle() # initialise the game and choose a random puzzle (a word) to guess

win = tk.Tk() # Create an instance of tkinter frame or window
win.geometry("500x800") # Set the size of the window

# The grid consists of 6 x 5 label widgets
for r in range(6):
    tk.Grid.rowconfigure(win, r, weight=1)
    l = []
    for c in range(5):
        tk.Grid.columnconfigure(win, c, weight=1)
        lab = tk.Label(text='', font=("Arial Bold", 20), fg='white', bg='white',  borderwidth=1, relief="solid")
        lab.grid(row=r, column=c, sticky='nesw')
        l.append(lab)
    labels.append(l)
        
entry = tk.Entry(width=30)
entry.grid(row=6, columnspan=5, sticky='nsew')
entry.bind('<Return>', on_enter)

win.mainloop()
```
